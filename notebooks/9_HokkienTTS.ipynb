```python
from transformers import VitsModel, AutoTokenizer
import torch
from IPython.display import Audio

model = VitsModel.from_pretrained("facebook/mms-tts-nan")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-nan")
```

```python
text = "Li chia̍h-pá bô"
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform

Audio(output, rate=model.config.sampling_rate)
```

```python
from scipy.io.wavfile import write
# Define output filename
output_filename = "hokkien_tts.wav"

# Save as a WAV file
write(output_filename, rate=model.config.sampling_rate, data=output.squeeze().numpy())
```
