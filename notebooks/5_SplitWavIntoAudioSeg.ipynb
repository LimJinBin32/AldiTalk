```python
# Unzip the wav files
import os
import zipfile

zipped_wav_directory = "/content/drive/MyDrive/NYP/Year2/AI Engineering Project/AIEP [shared]/wav" # YOUR WAV DIRECTORY HERE
output_directory = "/content/wav"
os.makedirs(output_directory, exist_ok=True)

for item in os.listdir(zipped_wav_directory):
    if item.endswith(".zip"):
        zip_path = os.path.join(zipped_wav_directory, item)
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            print(f"Extracting: {item}")
            zip_ref.extractall(output_directory)
        print(f"Extracted: {item} to {output_directory}")
```

```python
import json
from pydub import AudioSegment

# Paths to the large WAV file and JSON file
wav_file_path = "/content/wav/Towkay/“I Want To Be A Towkay” Episode 1《亲家，冤家做头家》第一集.wav"  # Replace with your large WAV file path
json_file_path = "/content/drive/MyDrive/NYP/Year2/AI Engineering Project/AIEP [shared]/ocr_cleaned/IWantToBeATowkay/“I Want To Be A Towkay” Episode 1《亲家，冤家做头家》第一集_ocr_cleaned.json"  # Replace with your JSON file path
output_directory = "/content/wav_segments"  # Replace with your desired output folder

# Create the output directory if it doesn't exist
import os
os.makedirs(output_directory, exist_ok=True)

# Load the JSON file
with open(json_file_path, 'r', encoding='utf-8') as json_file:
    transcript_data = json.load(json_file)

# Load the large WAV file
audio = AudioSegment.from_wav(wav_file_path)

# Process each segment from the JSON file
for segment in transcript_data:
    start_time = segment["start"] * 1000  # Convert seconds to milliseconds
    end_time = segment["end"] * 1000  # Convert seconds to milliseconds
    segment_filename = f"{segment['frame_name'].replace('.jpg', '')}.wav"

    # Extract the audio segment
    audio_segment = audio[start_time:end_time]

    # Save the audio segment as a new WAV file
    output_path = os.path.join(output_directory, segment_filename)
    audio_segment.export(output_path, format="wav")
    print(f"Segment saved: {output_path}")

print("Audio segmentation completed!")
```

```python
from IPython.display import Audio, display

# Iterate through the transcript data
for segment in transcript_data:
    start_time = segment["start"]
    end_time = segment["end"]
    transcript_en = segment.get("transcript_en", "No English transcript available")
    transcript_zh = segment.get("transcript_zh", "No Chinese transcript available")
    segment_filename = f"{segment['frame_name'].replace('.jpg', '')}.wav"

    # Construct the audio file path
    audio_path = os.path.join(output_directory, segment_filename)

    # Display transcripts and audio widget
    print(f"Start: {start_time}s | End: {end_time}s")
    print(f"English: {transcript_en}")
    print(f"Chinese: {transcript_zh}")
    display(Audio(audio_path))
```

```python
# time to process all the drama episodes
import os
import json
from pydub import AudioSegment

# Paths for input and output directories
wav_root_directory = "/content/wav"  # Root directory for large WAV files
json_root_directory = "/content/drive/MyDrive/NYP/Year2/AI Engineering Project/AIEP [shared]/ocr_cleaned"  # Root directory for JSON transcripts
output_root_directory = "/content/drive/MyDrive/NYP/Year2/AI Engineering Project/AIEP [shared]/wav_segments"  # Output directory for segmented WAV files

# Create the output root directory if it doesn't exist
os.makedirs(output_root_directory, exist_ok=True)

# Process each drama folder
for drama_folder in os.listdir(wav_root_directory):
    wav_drama_path = os.path.join(wav_root_directory, drama_folder)
    json_drama_path = os.path.join(json_root_directory, drama_folder)

    if os.path.isdir(wav_drama_path) and os.path.isdir(json_drama_path):
        print(f"Processing Drama: {drama_folder}")

        # Process each episode in the drama folder
        for wav_file in os.listdir(wav_drama_path):
            if wav_file.endswith(".wav"):
                # Construct paths for WAV file and corresponding JSON file
                wav_file_path = os.path.join(wav_drama_path, wav_file)
                episode_name = os.path.splitext(wav_file)[0]
                json_file_path = os.path.join(json_drama_path, f"{episode_name}_ocr_cleaned.json")

                if os.path.exists(json_file_path):
                    print(f"  Processing Episode: {episode_name}")

                    # Create output directory for the episode
                    episode_output_path = os.path.join(output_root_directory, drama_folder, episode_name)
                    os.makedirs(episode_output_path, exist_ok=True)

                    # Load the JSON transcript
                    with open(json_file_path, 'r', encoding='utf-8') as json_file:
                        transcript_data = json.load(json_file)

                    # Load the large WAV file
                    audio = AudioSegment.from_wav(wav_file_path)

                    # Process each segment from the JSON file
                    for segment in transcript_data:
                        start_time = segment["start"] * 1000  # Convert seconds to milliseconds
                        end_time = segment["end"] * 1000  # Convert seconds to milliseconds
                        segment_filename = f"{segment['frame_name'].replace('.jpg', '')}.wav"

                        # Extract the audio segment
                        audio_segment = audio[start_time:end_time]

                        # Save the audio segment
                        output_path = os.path.join(episode_output_path, segment_filename)
                        audio_segment.export(output_path, format="wav")
                        #print(f"    Segment saved: {output_path}")
                else:
                    print(f"  JSON file not found for Episode: {episode_name}")
    else:
        print(f"Skipping non-drama folder: {drama_folder}")

print("All dramas processed successfully!")
```
